# **1 one hot encoding using pandas**

```python
pd.get_dummies(df, columns=['fuel', 'owner'])
```

By default, `pd.get_dummies` returns columns with boolean values (`True` and `False`). If you prefer to have `0` and `1` for your dummy variables, you can specify the `dtype=int` argument in the `pd.get_dummies` function. Alternatively, you can convert the boolean columns to integers after creation.

```python
pd.get_dummies(df, columns=['fuel', 'owner'], dtype=int)
```

# **K-1 encoding**

```python
pd.get_dummies(df, columns=['fuel', 'owner'], dtype=int, drop_first=True)
```
